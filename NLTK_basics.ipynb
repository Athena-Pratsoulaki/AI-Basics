```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd


nltk.download('punkt')

nltk.download('stopwords')

nltk.download('wordnet')

nltk.download('omw-1.4')
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
print(data.loc[0])
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data['review_no_stopwords'][0]
```

```python
data['review_no_stopwords_no_punct'] = data.apply(lambda x : re.sub(r"[*]", "star", x['review_no_stopwords']), axis=1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data['review_no_stopwords_no_punct'].apply(lambda x : re.sub(r"([^\w\s])","", x))
```

```python
data.head()
```

```python
data['tokenized'] = data['review_no_stopwords_no_punct'].apply(lambda x: word_tokenize(x))
```

```python
data['tokenized'][0]
```

```python
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens : [lemmatizer.lemmatize(word) for word in tokens])
```

```python
data['lemmatized'][0]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply( lambda token : [ps.stem(word) for word in token]) 
```

```python
data['stemmed'][0]
```

```python
tokens_clean = sum(data['lemmatized'], [])
unigrams = (pd.Series(nltk.ngrams(tokens_clean,1)).value_counts())
print(unigrams)
```

```python
tokens_clean = sum(data['lemmatized'], [])
bigrams = (pd.Series(nltk.ngrams(tokens_clean,2)).value_counts())
print(bigrams)
```
